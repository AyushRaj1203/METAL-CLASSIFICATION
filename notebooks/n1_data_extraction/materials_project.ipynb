```python
from mp_api.client import MPRester
import csv

# Initialize MPRester with your API key
with MPRester("RWrpHlD6AcBElEP4vCRV8HIFh2HBp9oJ") as mpr:
    # Search for materials and specify fields to retrieve
    docs = mpr.materials.summary.search(fields=["material_id", "formula_pretty", "elements", "formation_energy_per_atom", "energy_above_hull", "volume", "band_gap", "efermi", "is_metal", "is_magnetic"])

    # Prepare data for CSV writing
    data_to_write = []
    for doc in docs:
        material_id = doc.material_id
        formula_pretty = doc.formula_pretty
        elements = doc.elements
        formation_energy_per_atom = doc.formation_energy_per_atom
        energy_above_hull = doc.energy_above_hull
        volume = doc.volume
        band_gap = doc.band_gap
        efermi = doc.efermi
        is_metal = doc.is_metal
        is_magnetic = doc.is_magnetic

        # Append fetched data to the list
        data_to_write.append({"Material ID": material_id, "Formula": formula_pretty, "Elements": elements, "Formation Energy": formation_energy_per_atom, "Energy Above Hull": energy_above_hull, "Volume": volume, "Band Gap": band_gap, "Fermi Level": efermi, "Is Metal": is_metal, "Is Magnetic": is_magnetic})

# Write the fetched data to a CSV file
filename = "material_data.csv"
csv_file_path = "../../data/d1_data_extraction/material_data.csv"
with open(csv_file_path, "w", newline="") as csvfile:
    fieldnames = ["Material ID", "Formula", "Elements", "Formation Energy", "Energy Above Hull", "Volume", "Band Gap", "Fermi Level", "Is Metal", "Is Magnetic"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write header
    writer.writeheader()

    # Write data rows
    for data_row in data_to_write:
        writer.writerow(data_row)

print("Data has been written to:", filename)
```
